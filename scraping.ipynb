```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs 
import time
```

```python
df = pd.read_csv('regional-us-weekly-latest.csv')
```

```python
new_header = df.iloc[0] #we will take the first row (row 0) as header , because it should be
df = df[1:] #get rid of the header (row 0)
df.columns = new_header #set the header as new_header
```

```python
df
```

```python
df['labels'] = None
```

```python
df
```

```python
for index in df.index:
    url = df.at[index, 'URL']
    r = requests.get(url)
    soup = bs(r.content)
    labels = [label.string for label in soup.find_all('p', class_="SYnLm8pwZls9sOX9AvJZ")]
    df.at[index, 'labels'] = labels
    print(index)
    time.sleep(1)
    
```

Now all we have to do to make lists of labels owed by major label with the said major label included obviously 

```python
#Universal Music Group 
universal = ['Universal Music Group', 'Universal Music', 'UMG', 'Interscope Records', 'Def Jam', 'Geffen Records', 'Interscope', 'EMI', 'Mercury Records', 'Taylor Swift','Virgin']

#Sony 
sony = ['Sony Music Entertainment', 'Columbia Records', 'Sony', 'RCA Records', 'AWAL Recordings', 'Epic Records', 'Columbia']

#Warner Bros 
warner = ['Warner Music Group', 'Warner', '300 Entertainment', 'Atlantic Recording', 'Atlantic Records', 'Sub Pop Records', 'Parlophone Records', 'Elektra Records']

#Disney 
disney = ['Walt Disney Records', 'Disney']

#Jay-Z
roc_nation = ['Roc Nation']
```

```python
df['major label'] = None
```

```python
def find_major_label(index) : 
    list_of_major_labels = [universal, sony, warner, disney, roc_nation]
    for major in list_of_major_labels : 
        for sub_label in major : 
            for i in range(len(df.at[index, 'labels'])):
                if sub_label in df.at[index, 'labels'][i]:
                    return major[0]

for index in df.index : 
    df.at[index, 'major label'] = find_major_label(index)
    df['major label'].fillna(value='Small Label/Independent', inplace=True)
    
```
